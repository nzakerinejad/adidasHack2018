```python
# my_office = (centLat, centLong)
my_office = (-10, 0)


degree_range = 80
extent = tilemapbase.Extent.from_lonlat(my_office[0] - degree_range, my_office[0] + degree_range,
                  my_office[1] - degree_range, my_office[1] + degree_range)
extent = extent.to_aspect(1.0)

extent
```

```python
fig, ax = plt.subplots(figsize=(16, 24), dpi=100)
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)

plotter = tilemapbase.Plotter(extent, t, width=1000)
plotter.plot(ax, t)

for values in zip(longMat, latMat):
# for values in [(0.0, 0.0)]:
    #print(values)  # print each tuples    print(values)  # print each tuples
    try:
        #if values in [(0.0, 0.0)]:
            #pass
        x, y = tilemapbase.project(*values)
        ax.scatter(x,y, marker=".", color="red", linewidth=2)
    except:
        pass
```
